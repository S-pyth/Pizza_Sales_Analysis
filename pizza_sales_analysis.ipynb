### KPI's

```python
# Calculate total revenue by summing the 'total_price' column
total_revenue = df['total_price'].sum()
# Calculate total number of pizzas sold by summing the 'quantity' column
total_pizzas_sold = df['quantity'].sum()
# Calculate total number of unique orders
total_orders = len(df['order_id'].unique())  
# Calculate average order value by dividing total revenue by number of orders
avg_order_value = total_revenue/total_orders
# Calculate average number of pizzas per order
avg_pizzas_per_order = total_pizzas_sold/total_orders

# Print the calculated metrics with formatted output
print(f"total_revenue: ${total_revenue:,.2f}")
print(f"total_pizzas_sold: {total_pizzas_sold:,}")
print(f"total_orders: {total_orders:,}")
print(f"avg_order_value: ${avg_order_value:.2f}")  
print(f"avg_pizzas_per_order: {avg_pizzas_per_order:.2f}")  
```

### Ingredient Analysis

```python
# Create a DataFrame of ingredient counts by:
# 1. Accessing the 'pizza_ingredients' column from the dataframe
# 2. Splitting each row's ingredients by comma
# 3. Exploding the lists to create one row per ingredient
# 4. Stripping whitespace from each ingredient
# 5. Counting occurrences of each ingredient
# 6. Resetting the index to make the ingredients a column
# 7. Renaming columns for clarity ('index' to 'ingrident' and 'pizza_ingredients' to 'count')
Ingredient = (
    df['pizza_ingredients']
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .reset_index()  # Changed from .reset.index() to .reset_index()
    .rename(columns={'index':'ingrident','pizza_ingredients':'count'})
)
print(Ingredient.head(15))
```

### Daily trends - Total Orders

```python
# Convert 'order_date' column to datetime format with day first (DD/MM/YYYY)
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

# Extract day name (Monday, Tuesday, etc.) from the order_date
df['day_name'] = df['order_date'].dt.day_name()

# Define the correct order of weekdays for sorting
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Convert day_name to categorical data type with specified order
df['day_name'] = pd.Categorical(df['day_name'], categories=weekday_order, ordered=True)

# Count unique orders for each day of the week
orders_by_day = df.groupby('day_name', observed=False)['order_id'].nunique()

# Create a bar plot of orders by day
ax = orders_by_day.plot(kind='bar', figsize=(8, 5), color='pink', edgecolor='black')

# Add title and labels to the plot
plt.title("Total orders by day of week")
plt.xlabel("Day of week")
plt.ylabel("Number of orders")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add value labels on top of each bar
for i, val in enumerate(orders_by_day):
    plt.text(i, val+20, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Adjust layout to prevent cutting off labels
plt.tight_layout()

# Display the plot
plt.show()
```

### Daily Trends-Total Revenue

```python
# Convert order_date column to datetime format with day first (European format)
df['order_date']=pd.to_datetime(df['order_date'],dayfirst=True)

# Extract day name (Monday, Tuesday, etc.) from order_date
df['day_name']=df['order_date'].dt.day_name()

# Define the correct order of weekdays for sorting
weekday_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

# Convert day_name to categorical data type with specified order
df['day_name']=pd.Categorical(df['day_name'],categories=weekday_order,ordered=True)

# Group data by day of week and sum the total price for each day
orders_by_day=df.groupby('day_name',observed=False)['total_price'].sum()

# Create a bar plot of total revenue by day of week
ax=orders_by_day.plot(kind='bar',figsize=(8,5),color='purple',edgecolor='black')

# Add title and labels to the plot
plt.title("Total Revenue by Day of Week")
plt.xlabel("Day of week")
plt.ylabel("Total Revenue")

# Rotate x-axis labels for better readability
plt.xticks(rotation= 45)

# Add value labels on top of each bar
for i , val in enumerate(orders_by_day):
    plt.text(i, val+20,str(val),ha='center',va='bottom',fontsize=9,fontweight='bold')

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```

### Hourly Trend-Total Orders

```python
# Convert order_time column to datetime format with specific time format
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')

# Extract the hour component from order_time and create a new column
df['order_hour'] = df['order_time'].dt.hour

# Group data by hour and count unique order IDs to get orders per hour
orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique()

# Create a bar plot of orders by hour with custom styling
ax = orders_by_hour.plot(kind='bar', figsize=(10, 5), color='maroon', edgecolor='black')

# Add title and axis labels to the plot
plt.title("Total Orders by Hour of Day")
plt.xlabel("Hour of Day (24-Hour Format)")
plt.ylabel("Number of Orders")

# Set x-axis tick rotation to 0 degrees (horizontal)
plt.xticks(rotation=0)

# Add data labels above each bar showing the exact number of orders
for i, val in enumerate(orders_by_hour):
    plt.text(i, val+5, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```

### Monthly Trends-Total Orders

```python
# Convert 'order_date' column to datetime format with day first (DD/MM/YYYY)
df['order_date']=pd.to_datetime(df['order_date'],dayfirst=True)

# Extract month names from the order dates
df['month_name']=df['order_date'].dt.month_name()

# Define the correct chronological order of months
month_order=["January","February","March","April","May","June","July","August","September","October","November","December"]

# Convert month_name to categorical data type with specified order for proper sorting
df['month_name']=pd.Categorical(df['month_name'],categories=month_order,ordered=True)

# Group data by month and count unique order IDs to get orders per month
orders_by_month=df.groupby('month_name',observed=False)['order_id'].nunique()

# Create a figure with specified size
plt.figure(figsize=(10,5))

# Create area plot with orange fill
plt.fill_between(orders_by_month.index,orders_by_month.values,color="orange",alpha=0.6)

# Add line plot with markers on top of area plot
plt.plot(orders_by_month.index,orders_by_month.values,color="black",linewidth=2,marker='o')

# Add title and axis labels
plt.title("Total Orders by Month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")

# Rotate x-axis labels for better readability
plt.xticks(rotation= 45)

# Add data labels above each point
for i , val in enumerate( orders_by_month):
    plt.text(i, val+20,str(val),ha='center',va='bottom',fontsize=9,fontweight='bold')

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the plot
plt.show()
```

### % of Sales by Category

```python
# Group data by pizza category and sum the total price for each category
category_sales = df.groupby('pizza_category')['total_price'].sum()

# Calculate the percentage of sales for each category
category_pct = category_sales/category_sales.sum()*100

# Create a figure with specified size
plt.figure(figsize=(7,7))

# Get a colormap for the pie chart
colors = plt.get_cmap('tab20').colors

# Create a donut chart showing percentage of sales by category
plt.pie(category_pct, 
        labels=category_pct.index, 
        autopct='%1.1f%%',  # Show percentage with 1 decimal place
        startangle=90,      # Start angle of the first slice
        colors=colors,      # Use the tab20 colormap
        wedgeprops={'edgecolor':'black', 'width':0.4})  # Add black edge and make it a donut chart

# Add a title to the chart
plt.title("Percentage of Sales by Pizza Category")

# Display the chart
plt.show()
```

### % Sales by Pizza Size & Category

```python
# Create a pivot table to summarize total sales by pizza category and size
sales_pivot=df.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0
)

# Convert the sales values to percentages of total sales
sales_pct=sales_pivot/sales_pivot.sum().sum()*100

# Create a heatmap visualization
plt.figure(figsize=(10,6))
sns.heatmap(sales_pct, annot=True, fmt=".1f", cmap="YlOrRd", linewidths=0.5)
plt.title("% of Sales by Pizza Category and Size")
plt.ylabel("pizza Category")
plt.xlabel("Pizza Size")
plt.show()
```
